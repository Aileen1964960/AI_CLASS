**1.	Text Analysis Tool**

```python
from google.colab import files
import re
from collections import Counter

# Uploaded the file
uploaded = files.upload()

# Show the file's content
with open('AI_1.txt', 'r') as file:
  content = file.read()

print("Contenido del archivo 'AI_1.txt':")
print(content)

# Realice the analysis of the file
try:
    # Count words, sentences, and paragraphs
    words = re.findall(r'\w+', content)
    sentences = re.split(r'[.!?]', content)
    paragraphs = content.split('\n\n')

    word_count = len(words)
    sentence_count = len(sentences) - 1
    paragraph_count = len(paragraphs)

    # Calculate average word length
    total = 0
    for word in words:
        total += len(word)
    average_word_length = total / word_count

    # Identify the most common words and their frequencies
    most_common_words = []

    max_word = None
    max_freq = 0

    while len(most_common_words) < 3:
        current_word = None
        current_freq = 0

        for word in words:
            if word != max_word:
                freq = words.count(word)
                if freq > current_freq:
                    current_word = word
                    current_freq = freq

        if current_word is not None:
            most_common_words.append((current_word, current_freq))
            max_word = current_word
        else:
            break

#Print the results
    print("\nAnalysis of the file:")
    print(f"Number of words: {word_count}")
    print(f"Number of sentences: {sentence_count}")
    print(f"Number of paragraphs: {paragraph_count}")
    print(f"Average word length: {average_word_length:.2f}")
    print("Most common words:")
    for word, frequency in most_common_words:
      print(f"{word}: {frequency}")

except Exception as e:
        print(f"Error at loaded'AI_1.txt'.")
```

**2.	Hangman Game**

```python
import random

# List of words to choose from
list = ["apple", "banana", "cherry", "date", "hello", "run", "grape", "honey", "kiwi", "lemon"]

# Function to choose a random word from the list
def choose_word():
    return random.choice(list)

# Function to play Hangman
def play_hangman():
    word_to_guess = choose_word()
    guessed_word = ['_'] * len(word_to_guess)
    attempts = 6

    print("Welcome to Hangman!")
    print(" ".join(guessed_word))

    while attempts > 0:
        guess = input("Guess a letter: ")

        if guess in word_to_guess:
            for i in range(len(word_to_guess)):
                if word_to_guess[i] == guess:
                    guessed_word[i] = guess

            print(" ".join(guessed_word))
        else:
            attempts -= 1
            print(f"Wrong guess, you have {attempts} attempts.")

        if "".join(guessed_word) == word_to_guess:
            print("You guessed the word correctly.")
            break

    if "".join(guessed_word) != word_to_guess:
        print(f"You're out of attempts. The word was '{word_to_guess}'.")

# Start the game
play_hangman()
```

**3.	Number Guessing Game**

```python
import random

# Generate a random number between 1 and 100
num = random.randint(1, 100)

# Initialize variables
guess = 0
attempts = 0

#Start guessing
while guess != num:
    try:
        guess = int(input("Enter your guess: "))
        attempts += 1

        if guess < num:
            print("Too low! Try a higher number.")
        elif guess > num:
            print("Too high! Try a lower number.")
        else:
            print(f"You guessed the number {num} in {attempts} attempts.")


    except ValueError:
        print("Please enter a valid number.")

```
